## Courbes de signaux et Multiplication de signaux

---

### Exemple :

On trace le signal $s(t)=3*cos(2\pi*f_{sig}*t)$ dans l'intervalle $t \in [0, 0.2]$ s

Avec $f_{sig}=20$

- On fixe la fréquence d'échantillonnage à $20*f_{sig}=20*20=400$ Hz

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fe = 400 # Fréquence d'échantillonnage
t = np.arange(0,0.2+(1/fe),1/fe) # Intervalle du temps

# Signal
f_sig = 20 # Fréquence du signal
A_sig = 3 # Amplitude du signal
s = A_sig*np.sin(2*np.pi*f_sig*t) # Formule du signal

# Courbes 
plt.figure(figsize=(10,6))
plt.subplot(311)
plt.plot(t,s)
plt.title('Signal')
plt.xlabel('Temps [s]')
plt.ylabel('Amplitude')
plt.grid()

plt.subplot(312)
plt.plot(t,s,'r')
plt.title('Signal')
plt.xlabel('Temps [s]')
plt.ylabel('Amplitude')
plt.grid()

plt.subplot(313)
plt.plot(t,s,'g')
plt.title('Signal')
plt.xlabel('Temps [s]')
plt.ylabel('Amplitude')
plt.grid()

plt.tight_layout()
plt.show()
```

## Modulations d'amplitude

La modulation d'amplitude, fait varier l'amplitude de la porteuse en fonction du signal modulant (message).

La porteuse (avant modulation) est un signal sinusoïdal de fréquence $f_p$ relativement élevée :

$$ x_{p}(t)=A_{p}\cos(\omega _{p}t)$$

- $A_p$ est l'amplitude maximale du signal
- $\omega_p = 2\pi f_p$ est la pulsation de la porteuse.

Le signal modulant est sinusoïdal, de fréquence $f_m$ nettement plus faible.

$$x_m(t)=A_m \cos(\omega_m t)$$

- $A_m$ est l'amplitude maximale du signal modulant
- $\omega_m=2\pi f_m$ est la pulsation du signal modulant.

Le signal modulé est donné par :

$$s(t)=A_{p}[1+m\cos(\omega _{m}t)]\cos(\omega _{p}t)$$

$m$ est l'indice de modulation

### Exercice 5 :

1. Tracer le signal $x_p(t)$
2. Tracer le signal $x_m(t)$
3. Tracer le signal modulé en amplitude $s(t)$ sur l'intervalle $t \in [0,2]$ s.

On donne :

- $x_p(t)= 1*\sin(2\pi*50*t)$
- $x_m(t)= 0.5*\sin(2\pi*2*t)$

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

fe = 400 # Fréquence d'échantillonnage
t = np.arange(0,0.2+(1/fe),1/fe) # Intervalle du temps

# Signal
f_sig = 20 # Fréquence du signal
A_sig = 3 # Amplitude du signal
s = A_sig*np.sin(2*np.pi*f_sig*t) # Formule du signal


fig = make_subplots(rows=2, cols=1, shared_xaxes=True)
fig.add_trace(go.Scatter(x=t,y=s,name="Signal1"), row=1, col=1)
fig.update_xaxes(title_text="Temps [s]", type="linear", row=1, col=1)
fig.update_yaxes(title_text="Amplitude", type="linear", row=1, col=1)

fig.add_trace(go.Scatter(x=t,y=s,name="Signal2"), row=2, col=1)
fig.update_xaxes(title_text="Temps [s]", type="linear", row=2, col=1)
fig.update_yaxes(title_text="Amplitide", type="linear", row=2, col=1)

fig.update_layout(legend_title='<b> Légende : </b>')
#fig.update_layout(legend_orientation="h")
#fig.update_layout(legend=dict(x=-.1, y=1.2))
fig.show()
```
