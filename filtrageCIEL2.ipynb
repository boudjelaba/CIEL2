$T(z)=\dfrac{z+2}{z^2+z+1}$

```python
from numpy import *
from pylab import *
from scipy import signal
%matplotlib inline

# Déclaration de la fonction de transfert à partir des coefficients
# du numérateur et du dénominateur de la TF''''
T=signal.TransferFunction([1, 2],[1,1,1])

# Calcul des Zéros et des Pôles de la fonction de transfert T
z,p,k=signal.tf2zpk([1, 2],[1,1,1]) # Calcul des Zéros, Pôles et Gain
print("z=",z ,", p=",p ,", k=",k) # Affichage des Zéros, Pôles et Gain

# Diagramme des Zéros et Pôles
figure()
plot(z.real, z.imag,"o", label="Zéros") # Affichage des Zéros sur le diagramme
plot(p.real, p.imag,"*", label="Pôles") # Affichage des Pôles sur le diagramme
xlabel('Partie réelle') # Titre de l'axe des abscisses
ylabel('Partie imaginaire') # Titre de l'axe des ordonnées
title('Lieu des zéros et pôles') # Titre de la figure
axis([-2.5, 1, -1, 1]) # Limites des Axes ([x_min, x_max, y_min, y_max])
legend() # Affichage de la légende des courbes tracées (Pôles, Zéros)
grid(True) # Activation de la grille
show() #On montre le graphique
```
